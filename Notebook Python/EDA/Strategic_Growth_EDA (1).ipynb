```python
from scipy.stats import normaltest

# ==========================================
# D'Agostino-Pearson Normality Test
# ==========================================

num_cols = data.select_dtypes(include=['int64', 'float64']).columns

results = []

for col in num_cols:
    statistic, p_value = normaltest(data[col].dropna())

    results.append({
        "Column": col,
        "Statistic": statistic,
        "P-Value": p_value,
        "Normal Distribution": "Yes" if p_value > 0.05 else "No"
    })

normality_data = pd.DataFrame(results)

normality_data["Statistic"] = normality_data["Statistic"].round(2)
normality_data["P-Value"] = normality_data["P-Value"].apply(lambda x: f"{x:.5f}")

normality_data

print("=== D'Agostino-Pearson Normality Test ===\n")

for _, row in normality_data.iterrows():
    print(f"{row['Column']}")
    print(f"Statistic : {row['Statistic']}")
    print(f"P-value   : {row['P-Value']}")

    if row["Normal Distribution"] == "Yes":
        print("Conclusion: Data is normally distributed.\n")
    else:
        print("Conclusion: Data is NOT normally distributed.\n")
```

```python
import scipy.stats as stats
from scipy.stats import shapiro
print("=== UJI NORMALITAS SHAPIRO-WILK (SAMPLE 5.000 BARIS) ===")
for col in num_cols:
    # Mengambil sampel acak 5000 baris karena limitasi library shapiro
    sample_data = data[col].sample(5000, random_state=42)
    stat, p_value = shapiro(sample_data)
    
    # Penentuan keputusan berdasarkan alpha = 0.05
    status = "Berdistribusi NORMAL" if p_value > 0.05 else "TIDAK Normal"
    print(f"Kolom {col:18} -> p-value: {p_value:.5f} | Kesimpulan: {status}")
```

```python
# Flag per event type
data['Impression'] = (data['event_type'] == 'Impression').astype(int)
data['Click']      = (data['event_type'] == 'Click').astype(int)
data['Like']       = (data['event_type'] == 'Like').astype(int)
data['Comment']    = (data['event_type'] == 'Comment').astype(int)
data['Share']      = (data['event_type'] == 'Share').astype(int)
data['Purchase']   = (data['event_type'] == 'Purchase').astype(int)
# Engagement = like + comment + share
data['engagement'] = ((data['event_type'].isin(['Like','Comment','Share']))).astype(int)

# Datetime breakdown
data['hour']  = data['timestamp'].dt.hour
data['month'] = data['timestamp'].dt.month
data['week']  = data['timestamp'].dt.isocalendar().week.astype(int)

# Audience alignment flags
data['gender_match'] = (data['target_gender'] == data['user_gender']).astype(int)
data['age_match']    = (data['target_age_group'] == data['age_group']).astype(int)
```

### Funnel Analysis

```python
# ── STEP 5: FUNNEL ANALYSIS (REVISED) ────────────────────────

funnel_order = ['Impression','Click','Like','Comment','Share','Purchase']
funnel = data['event_type'].value_counts().reindex(funnel_order).reset_index()
funnel.columns = ['event_type','count']

# Rate terhadap impression
base = funnel.loc[funnel['event_type']=='Impression','count'].values[0]
funnel['Rate_vs_Impression'] = (funnel['count'] / base * 100).round(2)

print(funnel)

# Metrik utama
impressions = funnel.loc[funnel['event_type']=='Impression','count'].values[0]
clicks      = funnel.loc[funnel['event_type']=='Click','count'].values[0]
engagements = data['engagement'].sum()
purchases   = funnel.loc[funnel['event_type']=='Purchase','count'].values[0]

print(f"\n📊 Funnel Metrics:")
print(f"CTR  (Click/Impression)      : {clicks/impressions*100:.2f}%")
print(f"ER   (Engagement/Impression) : {engagements/impressions*100:.2f}%")
print(f"PR   (Purchase/Click)        : {purchases/clicks*100:.2f}%")
print(f"ePR  (Purchase/Impression)   : {purchases/impressions*100:.2f}%")
```

#### ENGAGEMENT → PURCHASE PATH ANALYSIS 

```python
# Apakah user yang engage lebih cenderung purchase?
# Hitung per user_id: pernah engage? pernah purchase?

user_behavior = data.groupby('user_id').agg(
    ever_engaged  = ('engagement','max'),
    ever_clicked  = ('Click','max'),
    ever_purchased= ('Purchase','max')
).reset_index()

# Purchase rate: engaged users vs non-engaged
engaged_users     = user_behavior[user_behavior['ever_engaged']==1]
non_engaged_users = user_behavior[user_behavior['ever_engaged']==0]

print(f"Purchase rate (engaged users)     : {engaged_users['ever_purchased'].mean()*100:.2f}%")
print(f"Purchase rate (non-engaged users) : {non_engaged_users['ever_purchased'].mean()*100:.2f}%")
```

```python
# Gender match rate per event type
gender_by_event = data.groupby('event_type')['gender_match'].mean() * 100
print("Gender Match Rate per Event Type (%):\n", gender_by_event.round(1))

# Age match rate per event type
age_by_event = data.groupby('event_type')['age_match'].mean() * 100
print("\nAge Match Rate per Event Type (%):\n", age_by_event.round(1))

# Crosstab: target_age_group vs age_group
age_align = pd.crosstab(data['target_age_group'], data['age_group'], normalize='index') * 100
print("\nAge Group Alignment Matrix (%):\n", age_align.round(1))
```

### BUDGET EFFICIENCY PER CAMPAIGN 

```python
campaign_perf = data.groupby('campaign_id').agg(
    total_budget  = ('total_budget','first'),
    duration_days = ('duration_days','first'),
    impressions   = ('Impression','sum'),
    clicks        = ('Click','sum'),
    engagements   = ('engagement','sum'),
    purchases     = ('Purchase','sum'),
).reset_index()

campaign_perf['CTR'] = (campaign_perf['clicks'] / campaign_perf['impressions'].replace(0,np.nan) * 100).round(2)
campaign_perf['ER']  = (campaign_perf['engagements'] / campaign_perf['impressions'].replace(0,np.nan) * 100).round(2)
campaign_perf['PR']  = (campaign_perf['purchases'] / campaign_perf['clicks'].replace(0,np.nan) * 100).round(2)
campaign_perf['Cost_per_Purchase'] = (
    campaign_perf['total_budget'] / campaign_perf['purchases'].replace(0,np.nan)
).round(2)

print(campaign_perf.sort_values('total_budget', ascending=False))
```

#### LABEL ENCODING (Korelasi)

```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
df_encoded = data.copy()
le = LabelEncoder()

encode_cols = ['day_of_week','time_of_day','event_type','ad_platform','ad_type',
               'target_gender','target_age_group','target_interests',
               'user_gender','age_group','country','interests']

for col in encode_cols:
    df_encoded[col + '_enc'] = le.fit_transform(df_encoded[col].astype(str))

num_cols = [c for c in df_encoded.columns if c.endswith('_enc')] + \
           ['user_age','duration_days','total_budget','gender_match','age_match',
            'Click','engagement','Purchase']

corr = df_encoded[num_cols].corr()

plt.figure(figsize=(16,13))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, linewidths=0.5, annot_kws={"size":7})
plt.title('Correlation Matrix — All Features')
plt.tight_layout()
plt.savefig('correlation_matrix.png', dpi=150)
plt.show()
```
